* la méthode de corrélation de Kendall mesure la corrélation de rang entre deux variables, étant particulièrement résistante aux valeurs aberrantes et employée dans des situations où la distribution des données peut ne pas être normale, le coefficient de corrélation de Kendall étant fondé sur la concordance et la discordance des paires de rangs entre ces variables.
* La méthode de corrélation de Spearman mesure la corrélation de rang entre deux variables, elle est robuste aux valeurs aberrantes et est couramment utilisée lorsque les données ne suivent pas une distribution normale. Le coefficient de corrélation de Spearman évalue la monotonie de la relation entre deux variables en utilisant les valeurs réelles des rangs.

2.

Les méthodes de corrélation de Pearson, de Kendall et de Spearman évaluent la relation entre deux variables, chacune offrant une perspective différente. La corrélation de Pearson mesure la corrélation linéaire, adaptée aux données qui suivent une distribution normale. La corrélation de Kendall, basée sur la concordance et la discordance des paires de rangs, est robuste aux valeurs aberrantes et appropriée pour des données non normalement distribuées. La corrélation de Spearman, similaire à celle de Kendall mais utilisant les valeurs réelles des rangs, évalue la monotonie de la relation entre les variables et est également robuste aux valeurs aberrantes.
